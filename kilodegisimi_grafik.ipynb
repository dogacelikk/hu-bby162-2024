```python
import pandas as pd
import matplotlib.pyplot as plt

# Veriyi yükleme
veriSeti = pd.read_csv('weight_change_dataset.csv')

# Cinsiyete göre kilo değişim ortalamasını hesaplama
cinsiyet_kilo_degisim = veriSeti.groupby('Gender')['Weight Change (lbs)'].mean()

# Grafik boyutunu ayarlama
plt.figure(figsize=(15, 7))

# Çubuk grafik çizme
cinsiyet_kilo_degisim.plot(kind='bar', color=['cornflowerblue', 'purple'])
plt.xlabel('Cinsiyet')
plt.ylabel('Ortalama Kilo Değişimi (lbs)')
plt.title('Cinsiyete Göre Ortalama Haftalık Kilo Değişimi')
plt.xticks(rotation=0)
# Y-ekseni aralığını sıfırdan başlatma
plt.ylim(bottom=0)

# Grafiği gösterme
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Veriyi yükleme
veri = pd.read_csv('weight_change_dataset.csv')
# Günlük kalorik fazlayı katılımcının kilosuna oranlama
veri['Kalorik Fazla/Ağırlık Oranı'] = veri['Daily Caloric Surplus/Deficit'] / veri['Current Weight (lbs)']
# Grafik boyutunu ayarlama
plt.figure(figsize=(20, 7))
# Her katılımcı için kalorik fazla/ağırlık oranını çubuk grafik olarak çizme
plt.bar(veri['Participant ID'], veri['Kalorik Fazla/Ağırlık Oranı'], color='purple')
plt.xlabel('Katılımcı ID')
plt.ylabel('Kilo Başına Kalorik Fazla Oranı')
plt.title('Günlük Kalorik Fazlanın Başlangıç Ağırlığına Göre Oranı')
plt.xticks(rotation=90)

# Grafiği gösterme
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Veriyi yükleme
veri = pd.read_csv('weight_change_dataset.csv')
# Kilo değişiminin bazal metabolizma hızına (BMR) oranını hesaplama
veri['Kilo Değişimi/BMR Oranı'] = veri['Weight Change (lbs)'] / veri['BMR (Calories)']
# Grafik boyutunu ayarlama
plt.figure(figsize=(20, 7))
# Her katılımcı için kilo değişimi/BMR oranını çubuk grafik olarak çizme
plt.bar(veri['Participant ID'], veri['Kilo Değişimi/BMR Oranı'], color='cornflowerblue')
plt.xlabel('Katılımcı ID')
plt.ylabel('Kilo Değişiminin BMR\'ye Oranı')
plt.title('Kilo Değişimi ve Bazal Metabolizma Hızı (BMR) Oranı')
plt.xticks(rotation=90)
# Y-ekseni aralığını sıfırdan başlatma
plt.ylim(bottom=0)
# Grafiği gösterme
plt.show()


```
