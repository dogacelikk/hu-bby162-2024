```python
import pandas as pd
import matplotlib.pyplot as plt

veriSeti = pd.read_csv('mudahale-edilen-sokak-hayvanlari.csv', delimiter=';')

veri_2020 = veriSeti[veriSeti['YIL'] == 2020]
plt.figure(figsize=(10, 6))
plt.bar(veri_2020['AY'], veri_2020['TOPLAM_ADET'], color='red')
plt.xlabel('Aylar')
plt.ylabel('Müdahale Edilen Hayvan Sayısı')
plt.title('2020 Yılında Aylara Göre Müdahale Edilen Hayvan Sayısı')
plt.xticks(rotation=0)
plt.show()
```
